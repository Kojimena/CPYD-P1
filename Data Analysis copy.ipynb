```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Definir la ruta de la carpeta donde se encuentran los archivos
folder_path = 'iter'

# Listar todos los archivos en la carpeta
file_path = "./iter.log"

# Leer los datos del archivo
df = pd.read_csv(file_path, delim_whitespace=True, header=None, names=["creation_time", "display_time", "fps", "figures"])

# Eliminar las últimas 5 filas
df = df[:-5]

# en la primera fila viene la configuración de la simulación (N, EXPLOSION_FRAMES, SCREEN_WIDTH, SCREEN_HEIGHT)
config = df.iloc[0]
c = {
    "N": int(config["creation_time"]),
    "EXPLOSION_FRAMES": int(config["display_time"]),
    "SCREEN_WIDTH": int(config["fps"]),
    "SCREEN_HEIGHT": int(config["figures"])
}

df = df[1:]
```

```python
print(c)
df.head()
```

```python
# AGARRAR LAS PRIMERAS 2500 FILAS
df = df[:180]
```

```python
# Calcular los promedios
average_creation_time = df["creation_time"].mean()
average_display_time = df["display_time"].mean()
average_fps = df["fps"].mean()
average_figures = df["figures"].mean()

# Obtener los valores máximos y mínimos
max_creation_time = df["creation_time"].max()
min_creation_time = df["creation_time"].min()
max_display_time = df["display_time"].max()
min_display_time = df["display_time"].min()
max_fps = df["fps"].max()
min_fps = df["fps"].min()
max_figures = df["figures"].max()
min_figures = df["figures"].min()

# Crear las gráficas
plt.figure(figsize=(12, 12))

# Graficar tiempo de creación y despliegue
plt.subplot(3, 1, 1)
plt.plot(df.index, df["creation_time"], label="Creation Time", marker='o')
plt.plot(df.index, df["display_time"], label="Display Time", marker='x')
plt.axhline(y=average_creation_time, color='red', linestyle='--', label=f'Avg Creation Time: {average_creation_time:.2f}')
plt.axhline(y=average_display_time, color='blue', linestyle='--', label=f'Avg Display Time: {average_display_time:.2f}')
plt.axhline(y=max_creation_time, color='orange', linestyle='-', label=f'Max Creation Time: {max_creation_time:.2f}')
plt.axhline(y=min_creation_time, color='cyan', linestyle='-', label=f'Min Creation Time: {min_creation_time:.2f}')
plt.axhline(y=max_display_time, color='magenta', linestyle='-', label=f'Max Display Time: {max_display_time:.2f}')
plt.axhline(y=min_display_time, color='yellow', linestyle='-', label=f'Min Display Time: {min_display_time:.2f}')
plt.title(f"Creation Time and Display Time per Test ({c})")
plt.xlabel("Test Number")
plt.ylabel("Time (ms)")
plt.legend()

# Graficar FPS
plt.subplot(3, 1, 2)
plt.plot(df.index, df["fps"], label="FPS", color='green', marker='s')
plt.axhline(y=average_fps, color='purple', linestyle='--', label=f'Avg FPS: {average_fps:.2f}')
plt.axhline(y=max_fps, color='orange', linestyle='-', label=f'Max FPS: {max_fps:.2f}')
plt.axhline(y=min_fps, color='cyan', linestyle='-', label=f'Min FPS: {min_fps:.2f}')
plt.title(f"FPS per Test ({file_path})")
plt.xlabel("Test Number")
plt.ylabel("Frames per Second")
plt.legend()

# Graficar la cantidad de figuras
plt.subplot(3, 1, 3)
plt.plot(df.index, df["figures"], label="Figures", color='brown', marker='d')
plt.axhline(y=average_figures, color='purple', linestyle='--', label=f'Avg Figures: {average_figures:.2f}')
plt.axhline(y=max_figures, color='orange', linestyle='-', label=f'Max Figures: {max_figures:.2f}')
plt.axhline(y=min_figures, color='cyan', linestyle='-', label=f'Min Figures: {min_figures:.2f}')
plt.title(f"Figures per Test ({file_path})")
plt.xlabel("Test Number")
plt.ylabel("Number of Figures")
plt.legend()

# Mostrar las gráficas
plt.tight_layout()
plt.show()

# Imprimir los promedios y los valores máximos y mínimos
print(f'File: {file_path}')
print(f'Average Creation Time: {average_creation_time:.2f} ms')
print(f'Average Display Time: {average_display_time:.2f} ms')
print(f'Average FPS: {average_fps:.2f}')
print(f'Average Figures: {average_figures:.2f}')
print(f'Max Creation Time: {max_creation_time:.2f} ms')
print(f'Min Creation Time: {min_creation_time:.2f} ms')
print(f'Max Display Time: {max_display_time:.2f} ms')
print(f'Min Display Time: {min_display_time:.2f} ms')
print(f'Max FPS: {max_fps:.2f}')
print(f'Min FPS: {min_fps:.2f}')
print(f'Max Figures: {max_figures}')
print(f'Min Figures: {min_figures}')
print('-' * 40)
```
